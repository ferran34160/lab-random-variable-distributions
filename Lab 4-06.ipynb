```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt


customer_df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
numerical = customer_df.select_dtypes(np.number)
numerical.dtypes
```

```python
for column in numerical:
    sns.displot(x=numerical[column])
    plt.show()
```

```python
for column in numerical:
    sns.displot(numerical[column])
    sns.displot(numerical[column].apply(lambda x:np.log(x) if x>0 else x))
    plt.show()
```
